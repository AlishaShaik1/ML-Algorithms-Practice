KNN on iris data set

```python
from sklearn.datasets import load_iris
df=load_iris()
x=df.data
y=df.target
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
xt,xte,yt,yte=train_test_split(x,y,test_size=0.3)
print(xt)
print(xte)
print(yt)
print(yte)
```

```python
from sklearn.neighbors import KNeighborsClassifier
m=KNeighborsClassifier()
m.fit(xt,yt)
yp=m.predict(xte)
from sklearn.metrics import accuracy_score
print(accuracy_score(yte,yp))
```

KNN on Msuhroom Data set

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

df = pd.read_csv('/content/mushrooms.csv')

x = df.iloc[:, 1:].values
y = df.iloc[:, 0].values
new_mushroom = [['x','s','n','t','p','f','c','n','k','e',
                 'e','s','s','w','w','p','w','o','p','k','s','u']]
le = LabelEncoder()
y = le.fit_transform(y)

ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
x = ohe.fit_transform(x)
new_mushroom_encoded = ohe.transform(new_mushroom)
```

```python
from sklearn.model_selection import train_test_split
xt,xte,yt,yte=train_test_split(x,y,test_size=0.3)
print(xt)
print(xte)
print(yt)
print(yte)
```

```python
from sklearn.neighbors import KNeighborsClassifier
m=KNeighborsClassifier(n_neighbors=5)
m.fit(xt,yt)
yp=m.predict(xte)
from sklearn.metrics import accuracy_score
print(accuracy_score(yte,yp))
prediction = m.predict(new_mushroom_encoded)
result = le.inverse_transform(prediction)
print("Prediction:", result[0])
```

KNN on Titanic Data set

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/Titanic-Dataset.csv')
```

```python
X = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
y = df['Survived']
```

```python
X.loc[:, 'Age'] = X['Age'].fillna(X['Age'].median())
X.loc[:, 'Embarked'] = X['Embarked'].fillna(X['Embarked'].mode()[0])
```

```python
le_sex = LabelEncoder()
le_embarked = LabelEncoder()

X.loc[:, 'Sex'] = le_sex.fit_transform(X['Sex'])        # male/female → 0/1
X.loc[:, 'Embarked'] = le_embarked.fit_transform(X['Embarked'])
```

```python
xt, xte, yt, yte = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
scaler = StandardScaler()
xt = scaler.fit_transform(xt)
xte = scaler.transform(xte)
```

```python
model = KNeighborsClassifier(n_neighbors=5)
model.fit(xt, yt)
```

```python
yp = model.predict(xte)
print("Test Accuracy:", accuracy_score(yte, yp))
```

Multi Layer Neaural Network on Iris Data set

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
iris=load_iris()
X=iris.data
y=iris.target

xt,xte,yt,yte=train_test_split(X,y,test_size=00.3,random_state=42)
```

```python
scaler=StandardScaler()
xt=scaler.fit_transform(xt)
xte=scaler.transform(xte)
```

```python
model=MLPClassifier(hidden_layer_sizes=(10,),
activation='relu',
max_iter=1000,
random_state=42
)
```

```python
model.fit(xt,yt)
yp=model.predict(xte)
print("Accuracy:",accuracy_score(yte,yp))
```

```python
new_flower=[[5.1,3.5,1.4,0.2]]
new_flower_scaled=scaler.transform(new_flower)

prediction=model.predict(new_flower_scaled)
print("Predicted class:",iris.target_names[prediction[0]])
```

Heart Disease Prediction using Random Forest

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
X=df.drop(columns=['Heart Disease'])
y=df['Heart Disease']

xt,xte,yt,yte=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model=RandomForestClassifier(n_estimators=100,random_state=42)
```

```python
model.fit(xt, yt)
yp=model.predict(xte)
print("Accuracy:",accuracy_score(yte,yp))
print("\nClassification Report:\n",classification_report(yte,yp))
```

LOGISTIC REGRESSION MODEL ON HEART DISEASE

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv('/content/Heart_Disease_Prediction.csv')
X=df.drop(columns=['Heart Disease'])
y=df['Heart Disease']

xt,xte,yt,yte=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
scaler = StandardScaler()
xt = scaler.fit_transform(xt)
xte = scaler.transform(xte)
```

```python
model = LogisticRegression(
    max_iter=5000,        # important for large feature set
    solver='saga',        # best for many features
    penalty='l2',         # regularization
    n_jobs=-1
)
```

```python
model.fit(xt, yt)
yp = model.predict(xte)

print("Accuracy:", accuracy_score(yte, yp))
print("\nClassification Report:\n", classification_report(yte, yp))
```

Multi-Layer Neural Network On Heart Disease

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
xt, xte, yt, yte = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
scaler = StandardScaler()
xt = scaler.fit_transform(xt)
xte = scaler.transform(xte)
```

```python
mlp = MLPClassifier(
    hidden_layer_sizes=(128, 64),
    activation='relu',
    solver='adam',
    alpha=0.001,
    batch_size=32,
    learning_rate='adaptive',
    max_iter=1000,
    random_state=42,
    early_stopping=True
)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train_encoded = le.fit_transform(yt)
y_test_encoded = le.transform(yte)

mlp.fit(xt, y_train_encoded)
yp = mlp.predict(xte)

print("MLP Accuracy:", accuracy_score(y_test_encoded, yp))
print("\nClassification Report:\n", classification_report(y_test_encoded, yp))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(
    RandomForestClassifier(random_state=42),
    X, y,
    cv=5,
    scoring='accuracy'
)

print("CV Accuracy:", scores.mean())
```

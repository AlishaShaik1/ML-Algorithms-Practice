KNN on iris data set

```python
from sklearn.datasets import load_iris
df=load_iris()
x=df.data
y=df.target
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
xt,xte,yt,yte=train_test_split(x,y,test_size=0.3)
print(xt)
print(xte)
print(yt)
print(yte)
```

```python
from sklearn.neighbors import KNeighborsClassifier
m=KNeighborsClassifier()
m.fit(xt,yt)
yp=m.predict(xte)
from sklearn.metrics import accuracy_score
print(accuracy_score(yte,yp))
```

KNN on Msuhroom Data set

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

df = pd.read_csv('/content/mushrooms.csv')

x = df.iloc[:, 1:].values
y = df.iloc[:, 0].values
new_mushroom = [['x','s','n','t','p','f','c','n','k','e',
                 'e','s','s','w','w','p','w','o','p','k','s','u']]
le = LabelEncoder()
y = le.fit_transform(y)

ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
x = ohe.fit_transform(x)
new_mushroom_encoded = ohe.transform(new_mushroom)
```

```python
from sklearn.model_selection import train_test_split
xt,xte,yt,yte=train_test_split(x,y,test_size=0.3)
print(xt)
print(xte)
print(yt)
print(yte)
```

```python
from sklearn.neighbors import KNeighborsClassifier
m=KNeighborsClassifier(n_neighbors=5)
m.fit(xt,yt)
yp=m.predict(xte)
from sklearn.metrics import accuracy_score
print(accuracy_score(yte,yp))
prediction = m.predict(new_mushroom_encoded)
result = le.inverse_transform(prediction)
print("Prediction:", result[0])
```

KNN on Titanic Data set

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/Titanic-Dataset.csv')
```

```python
X = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
y = df['Survived']
```

```python
X['Age'].fillna(X['Age'].median(), inplace=True)
X['Embarked'].fillna(X['Embarked'].mode()[0], inplace=True)
```

```python
le_sex = LabelEncoder()
le_embarked = LabelEncoder()

X['Sex'] = le_sex.fit_transform(X['Sex'])        # male/female → 0/1
X['Embarked'] = le_embarked.fit_transform(X['Embarked'])
```

```python
xt, xte, yt, yte = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
scaler = StandardScaler()
xt = scaler.fit_transform(xt)
xte = scaler.transform(xte)
```

```python
model = KNeighborsClassifier(n_neighbors=5)
model.fit(xt, yt)
```

```python
yp = model.predict(xte)
print("Test Accuracy:", accuracy_score(yte, yp))
```

Single Layer Nearul Network on Titanic Data set

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('/content/Titanic-Dataset.csv')
```

```python
X=df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
y=df['Survived']
```

```python
X['Age'].fillna(X['Age'].median(),inplace=True)
X['Embarked'].fillna(X['Embarked'].mode()[0], inplace=True)
```

```python
le_sex=LabelEncoder()
le_emb=LabelEncoder()
X['Sex']=le_sex.fit_transform(X['Sex'])
X['Embarked'] = le_emb.fit_transform(X['Embarked'])
```

```python
xt,xte,yt,yte=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
scaler=StandardScaler()
xt=scaler.fit_transform(xt)
xte=scaler.transform(xte)

model=LogisticRegression()
model.fit(xt,yt)
yp=model.predict(xte)
print("Test Accuracy:",accuracy_score(yte,yp))
```

Multi Layer Neaural Network on Iris Data set

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
iris=load_iris()
X=iris.data
y=iris.target

xt,xte,yt,yte=train_test_split(X,y,test_size=00.3,random_state=42)
```

```python
scaler=StandardScaler()
xt=scaler.fit_transform(xt)
xte=scaler.transform(xte)
```

```python
model=MLPClassifier(hidden_layer_sizes=(10,),
activation='relu',
max_iter=1000,
random_state=42
)
```

```python
model.fit(xt,yt)
yp=model.predict(xte)
print("Accuracy:",accuracy_score(yte,yp))
```

```python
new_flower=[[5.1,3.5,1.4,0.2]]
new_flower_scaled=scaler.transform(new_flower)

prediction=model.predict(new_flower_scaled)
print("Predicted class:",iris.target_names[prediction[0]])
```
